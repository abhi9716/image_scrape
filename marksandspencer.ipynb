```
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get("https://www.marksandspencer.com/l/men/tops-tshirts-and-polos/tshirts#intid=gnav_men_clothing_Tshirts")
time.sleep(5)
driver.find_element_by_id('consent_prompt_submit').click()
```

```
links=[]
n=0
s="sdf"
try:
  while(s):
    # for i in range(0,50):
    #   # time.sleep(1)
    #   driver.execute_script("window.scrollTo(0, (document.body.scrollHeight));")
  
    s=driver.find_elements_by_css_selector('a.pagination__next.btn.btn--tertiary')
    # time.sleep(2)
    print(s)
    ActionChains(driver).move_to_element(s[0]).click(s[0]).perform()
    # s[0].click()
    time.sleep(2)
    a=driver.find_elements_by_css_selector('a.product')

    for d in a:
      # print(d.get_attribute('href'))
      links.append(d.get_attribute('href'))
    print(links[-1])
    n+=1
    print(n)
except:
  pass
```

```
links=(list(set(links)))
len(links)
```

```
from urllib.parse import urlparse
def save_img(prod,images):
  parent_dir="/content/drive/My Drive/zara-tops-data/mark_men/"
  path = os.path.join(parent_dir, prod.replace("/","-"))
  # print(path)
  if not os.path.isdir(path):
      os.mkdir(path)
  for i,l in enumerate(images):
    if(l.startswith('https')):
      t=urlparse(l)
      n=os.path.basename(t.path)
      with open(path +"/"+ str(i)+".jpg", 'wb') as handle:
        response = requests.get(l, stream=True)

        if not response.ok:
            print(response)

        for block in response.iter_content(1024):
            if not block:
                break

            handle.write(block)
```
