```
# you can change the product catagory link here
# copy link from website and paste it here
prod_url="https://en.zalando.de/mens-clothing-pullover-cardigans/"
```

```
# dir in drive where data will save
parent_dir="/content/drive/My Drive/tops-data/zalando"
# create folder where data of each catagory will save
folder_name=prod_url.split("/")[-2]
parent_path = os.path.join(parent_dir, folder_name.replace("/","-"))
if not os.path.isdir(parent_path):
  os.mkdir(parent_path)
print(parent_path)
```

```
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get(prod_url)
time.sleep(5)
driver.find_element_by_id('uc-btn-accept-banner').click()
```

```
links=[]
n=0
try:
  while True:
    # for i in range(0,50):
    #   # time.sleep(1)
    #   driver.execute_script("window.scrollTo(0, (document.body.scrollHeight));")

    s=driver.find_elements_by_css_selector('span.z-icon.z-icon-pagination-chevron-right-active.z-icon-small.z-icon-black')
    ActionChains(driver).move_to_element(s[0]).click(s[0]).perform()
    # s[0].click()
    time.sleep(2)
    # driver.implicitly_wait(10)
    a=driver.find_elements_by_css_selector('a.cat_imageLink-OPGGa')

    for d in a:
      # print(d.get_attribute('href'))
      links.append(d.get_attribute('href'))
    print(links[-1])
    n+=1
    print(n)
except:
  pass
```

```
links=list(set(links))
len(links)
```

```
from urllib.parse import urlparse
def save_img(prod,images,parent_path=parent_path):
    # parent_path=parent_path
    path = os.path.join(parent_path, prod.replace("/","-"))
    # print(path)
    if not os.path.isdir(path):
        os.mkdir(path)
    for i,l in enumerate(images):
            # if(l.startswith('https')):
              t=urlparse(l)
              n=os.path.basename(t.path)
              with open(path +"/"+ str(n), 'wb') as handle:
                  response = requests.get(l, stream=True)

                  if not response.ok:
                      print(response)

                  for block in response.iter_content(1024):
                      if not block:
                          break

                      handle.write(block)
```
